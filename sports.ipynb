```python
"""Найти самый популярный объект в выборке, объяснить почему.

Для начала разберемся в смысле предложенных метрик:

•	Lifetime Post Total Reach – количество уникальных пользователей, просмотревших пост;
•	Lifetime Post Total Impressions – количество показов (возможно несколько показов для одного человека);
•	Lifetime Engaged Users  and Lifetime Post Consumers - количество уникальных пользователей, которые кликнули пост. 
Engagement включает все клики приводящие к story. 
Consumption включает в себя все клики, независимо от того, приводят ли они к story или нет. 
В целом, метрики очень похожи, но для анализа популярности поста важнее количество любых кликов, 
тк они учитывают любые взаимодействия пользователя с контентом, 
включае те, которые не оставляют следов в story; 
Link https://brnrd.me/posts/engagement-vs-consumption-the-facebook-dilemma 
•	Lifetime Post Consumptions – количество всех кликов;
•	Lifetime Post Impressions by people who have liked your Page – количество показов пользователям, которые лайнкули страницу 
•	Lifetime Post reach by people who like your Page - количество показов уникальным пользователям, которые лайкнули страницу
•	Lifetime People who have liked your Page and engaged with your post - количество пользователей, которые лайкнули страницу и взаимодейстовали с постом. 

По моему мнению, популярность можно измерить количеством уникальных пользователей, 
которые просмотрели пост и, в то же время, пост их настолько заинетересовал,
что они начали взаимодействие с ним - начали его кликать.
Поэтому, как меру популярности необходимо выбрать Lifetime Post Consumers.
Выберем топ пост с самой высокой метрикой LPC:"""

df.sort_values(by = 'Lifetime Post Consumers', ascending=False).head(1)
```
